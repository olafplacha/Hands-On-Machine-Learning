```python
import tensorflow as tf
from sklearn.datasets import fetch_mldata
import numpy as np
```

```python
def shuffle_dataset(X, y):
    
    random_idx = np.random.permutation(X.shape[0])
    
    return X[random_idx], y[random_idx]
```

```python
def train_test_split(X, y, ratio=0.8):
    
    X_shuffled, y_shuffled = shuffle_dataset(X, y)
    X_train, y_train = X_shuffled[:int(ratio*X.shape[0])], y_shuffled[:int(ratio*X.shape[0])]
    X_test, y_test = X_shuffled[int(ratio*X.shape[0]):], y_shuffled[int(ratio*X.shape[0]):]
    
    return X_train, y_train, X_test, y_test
```

```python
X_train, y_train, X_test, y_test = train_test_split(fetch_mldata('MNIST original').data, fetch_mldata('MNIST original').target)
X_train = X_train/255
X_test = X_test/255
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

```python
feature_columns = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300,100], n_classes=10, feature_columns=feature_columns)
dnn_clf.fit(x=X_train, y=y_train, batch_size=50, steps=40000)
```

```python
dnn_clf.evaluate(X_test, y_test)
```
